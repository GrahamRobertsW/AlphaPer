```python
import pandas as pd
import psycopg2 as p2
```

```python
con=p2.connect("dbname='stars' user='postgres' host='localhost'")
cur=con.cursor()
```

```python
cur.execute("select ra, dec, radius, spt, dm, pmra, pmdec, spm, mem, memp, mem1 from alejandroslist;")
al=pd.DataFrame(cur.fetchall(),columns=['ra','de','radius','spt','dm','pmra','pmde', 'spm', 'mem', 'memp', 'mem1'])
```

```python
cur.fetchall()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(al['mem'],bins=10,log=True,color=(1,1,1))
plt.xlabel('membership')
plt.ylabel('count')
plt.title("Histogram of alejandro's membership probabilities")
plt.show()
```

```python
x=[]
y=[]
pmx=[]
pmy=[]
cols=[]
a=[]
for i in range(len(al)):
    pmx=pmx+[al['pmra'][i]]
    pmy=pmy+[al['pmde'][i]]
    cols=cols+[(1-.01*al['mem'][i],0,.01*al['mem'][i])]
    al['col'][i]
    x=x+[al['ra'][i]]
    y=y+[al['de'][i]]
    a=a+[0.5+0.005*al['mem'][i]]
```

```python
plt.scatter(x,y,color=cols,marker='o',alpha=0.4)
plt.xlabel('Right Ascension')
plt.ylabel('Declination')
plt.show()
```

```python
plt.scatter(pmx,pmy,color=cols,marker='o',alpha=0.4)
plt.xlabel('pmRA')
plt.ylabel('pmDE')
plt.show()
```

```python
al['col']="#{0:02X}{1:02X}{2:02X}".format((int((1-0.01*float(al['mem']))*255)),0,int((0.01*float(al['mem']))*255))
```
